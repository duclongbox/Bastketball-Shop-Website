```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
item_list = []
base_url = 'https://stockx.com/shoes?page'
totalPage = 25
for page in range(1,totalPage+1):
  url = f'{base_url}{page}'
  html = requests.get(url)
  soup = BeautifulSoup(html.content,'html.parser')
  price_element= soup.find_all('p',class_='chakra-text css-nsvdd9')
  p_element = soup.find_all('p', class_='chakra-text css-3lpefb')
  for i in range(1,len(p_element)):
    item_name = p_element[i].text
    item_price = price_element[i].text
    data= {
        'name': item_name,
        'price': item_price
    }
    item_list.append(data)

```

```python
goodread_pf = pd.DataFrame(item_list)
goodread_pf.to_csv('stockxTestData.csv',index=False)
```
